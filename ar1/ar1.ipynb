```python
import neatnik
import numpy  as np
import pickle as p
import matplotlib.pyplot as plt
```

```python
with open('data.p', 'rb') as file:
    x = p.load(file)
xx = np.sort(x, axis=1)
```

```python
plt.plot(x.flatten())
```

```python
# Loads constrained result.
with open('./constrained/ar1.p', 'rb') as file:
    constrained = p.load(file)

# Loads unconstrained result.
with open('./unconstrained/ar1.p', 'rb') as file:
    unconstrained = p.load(file)

# Loads regularized result.
with open('./regularized/ar1.p', 'rb') as file:
    regularized = p.load(file)
```

```python
# Converts results into networks.
constrained = neatnik.Organism(constrained)
unconstrained = neatnik.Organism(unconstrained)
regularized = neatnik.Organism(regularized)
```

```python
plt.scatter(x[:,:-1,:],x[:,1:,:], color='black', s=5, alpha=0.2)
plt.plot(x[:,:-1,:].flatten(), (0.5 + 0.85*x[:,:-1,:]).flatten(), color='red', label='Exact')
plt.plot(xx.flatten(), constrained.react(xx).flatten(), color='orange', label='Constrained')
plt.plot(xx.flatten(), unconstrained.react(xx).flatten(), color='blue', label='Unconstrained')
plt.plot(xx.flatten(), regularized.react(xx).flatten(), color='green', label='Regularized')
plt.legend()
```
